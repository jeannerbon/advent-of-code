```python
#si la longueur du nombre est pair alors on verifie que les deux parties de ce nombre ne sont pas identiques

line = "10-15,11-22"

total = 0

for part in line.split(","):
    a, b = map(int, part.split("-"))

    for n in range(a, b + 1): # pour les nombres entre 10 et 15
        s = str(n)

        print("\nNombre:", n, "| s =", s, "| longueur =", len(s))

        if len(s) % 2 == 0:  # longueur paire (ex : 10)
            mid = len(s) // 2

            left = s[:mid] # 1
            right = s[mid:] # 0

            print("  longueur paire ‚Üí mid =", mid)
            print("  gauche  = s[:mid]  =", left)
            print("  droite  = s[mid:]  =", right)

            if left == right:
                print("  üëâ INVALID (X == X)")
                total += n
            else:
                print("  üëâ valide (X != X)")
        else:
            print("  longueur impaire ‚Üí automatiquement valide")

print("\nTOTAL =", total)
```

### part 2

Un bloc est invalide si il est r√©p√©t√© AU MOINS 2 FOIS

```python
line = open("/Users/jeanne/Documents/Challenge python 2025/input_day2_part2.txt").read()#.strip() permet d'enlever ou de suprimer les espaces devant ou derri√®re \n ou espqce

total = 0

for part in line.split(","):
    if not part:
        continue

    a, b = map(int, part.split("-"))

    for n in range(a, b + 1):
        s = str(n)
        L = len(s)

        #d = taille du bloc candidat, peut √™tre impaire ou pair et on regarde pour chaque taille possible si il y a occurence de ce bloc
        for d in range(1, L // 2 + 1):
            if L % d != 0:
                continue

            block = s[:d]
            if block * (L // d) == s:
                total += n
                break

print(total)
```

```python
def best_12_digits(line):
    digits = list(line)
    remove = len(digits) - 12   # combien on doit enlever
    stack = []

    for d in digits:
        while remove > 0 and len(stack) > 0 and stack[-1] < d:
            stack.pop() #supprime le dernier √©l√©ment de la liste stack
            remove = remove - 1
        stack.append(d)# ajoute le chiffre plus grand

    # si on doit encore enlever (cas o√π la suite est d√©j√† d√©croissante)
    if remove > 0:
        stack = stack[:-remove]# enl√®ve les derniers chiffres √† la fin en fonction de combien de fois la boucle while a fonctionn√©
    # on garde exactement 12 chiffres
    result = "".join(stack[:12])#pour convertir la chaine de caract√®re en nombre
    return int(result)

def total_joltage_12(text):
    total = 0
    for line in text.splitlines():
        if line.strip(): #le strip enl√®ve les espaces
            total = total + best_12_digits(line)
    return total

# Exemple (doit donner 3121910778619)
data = """987654321111111
811111111111119
234234234234278
818181911112111
"""


data = open('/Users/jeanne/Documents/Challenge python 2025/input_day3_part2.txt').read()
print(total_joltage_12(data))
```

```python
path = '/Users/jeanne/Documents/Challenge python 2025/input_day5.txt'


def count_fresh_ids(path):

    # 1) Lire le fichier
    with open(path, "r") as f:
        text = f.read()

    ranges_part, ids_part = text.split("\n\n", 1) #on coupe l√† ou il y a une ligne vide 1 fois, donc le resultat donne 2 morceaux

    # 3) Lire les ranges
    ranges = [] #on cr√©e une liste de tuples (,) : les tuples sont des donnes non modifiables comparees aux listes
    for line in ranges_part.splitlines():
        start, end = line.split("-") # on coupe les range en 2 (+ petit et plus grand)
        ranges.append((int(start), int(end)))

    print(ranges)

    # 4) Lire les ids
    ids = []#liste contenant les ids 
    for line in ids_part.splitlines():
        ids.append(int(line))

    # 5) V√©rifier quels ids sont frais : Un id est consid√©r√© comme ‚Äúfrais‚Äù s‚Äôil appartient √† au moins un tuple (start, end)
    fresh = 0
    for x in ids:
        for start, end in ranges:
            if start <= x <= end:
                fresh += 1
                break   # on s‚Äôarr√™te d√®s qu‚Äôun range correspond

    return fresh
```

```python
path = '/Users/jeanne/Documents/Challenge python 2025/input_day5.txt'


def count_fresh_from_ranges(path):
    with open(path, "r") as f:
        text = f.read().strip()

    ranges_part = text.split("\n\n", 1)[0]

    ranges = []
    for line in ranges_part.splitlines():
        a, b = line.split("-")
        ranges.append((int(a), int(b)))

    # 1) Trier les ranges
    ranges.sort()

    # 2) Fusionner les ranges qui se chevauchent
    total = 0
    cur_start, cur_end = ranges[0]
    
    for a, b in ranges[1:]:
        if a <= cur_end + 1:          # chevauchement
            cur_end = max(cur_end, b)
        else:                         # nouveau bloc
            total += cur_end - cur_start + 1
            cur_start, cur_end = a, b

    # 3) Ajouter le dernier intervalle
    total += cur_end - cur_start + 1

    return total



print(count_fresh_from_ranges(path))

```

```python
def extract_ints(s):
    nums, cur = [], ""
    for ch in s:
        if ch.isdigit():
            cur += ch
        else:
            if cur:
                nums.append(int(cur))
                cur = ""
    if cur:
        nums.append(int(cur))
    return nums


# IMPORTANT : on pad toutes les lignes √† la m√™me largeur
w = max(len(l) for l in lines) if lines else 0
grid = [l.ljust(w) for l in lines]


total = 0
start = 0


for c in range(w + 1):
    # colonne vide (sur TOUTES les lignes), ou fin
    if c == w or all(row[c] == ' ' for row in grid):
        if start < c:
            block = [row[start:c] for row in grid]
            op = '+' if '+' in block[-1] else '*'

            nums = []
            for row in block[:-1]:
                nums += extract_ints(row)

            if op == '+':
                total += sum(nums)
            else:
                p = 1
                for x in nums:
                    p *= x
                total += p

        # on repart √† la colonne suivante (mais sans bug sur multi-espaces)
        start = c + 1

print("Grand total =", total)
```
